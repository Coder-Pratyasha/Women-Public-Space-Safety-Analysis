# Unsupervised Modeling

This notebook applies clustering techniques to group public places based on safety-related features.

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, AgglomerativeClustering

place_data = pd.read_csv("data/feature_data.csv")
```

## Feature Scaling

Features are standardized using StandardScaler to ensure equal contribution during clustering.

```python
scaler = StandardScaler()
X = scaler.fit_transform(
    place_data[['Infrastructure_Score','Perception_Score']]
)
```

## K-Means Clustering

K-Means is applied to partition public places into clusters based on safety characteristics.

```python
kmeans = KMeans(n_clusters=3, random_state=42)
place_data['kmeans'] = kmeans.fit_predict(X)
```

## Hierarchical Clustering

Hierarchical clustering (Ward and Average linkage) is used to understand relationships between clusters.

```python
ward = AgglomerativeClustering(n_clusters=3, linkage='ward')
place_data['ward'] = ward.fit_predict(X)
```

```python
avg = AgglomerativeClustering(n_clusters=3, linkage='average')
place_data['avg'] = avg.fit_predict(X)
```

## Ensemble Clustering

Final clusters are obtained using majority voting across different clustering methods to improve robustness.

```python
cluster_cols = ['kmeans','ward','avg']
place_data['final_cluster'] = place_data[cluster_cols].mode(axis=1)[0]

print(place_data[['Type_of_Public_Place','kmeans','ward','avg','final_cluster']])
```

```python
place_data.to_csv("data/clustered_data.csv", index=False)
```
