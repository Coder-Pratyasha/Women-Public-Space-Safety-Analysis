# Feature Engineering

This notebook focuses on creating meaningful features that represent safety more effectively.

```python
import pandas as pd

df = pd.read_csv("data/processed_data.csv")
```

# Feature Creation
New features such as Infrastructure Score and Perception Score are created by combining relevant variables.

# Purpose of Features

- Infrastructure Score represents availability of safety facilities  
- Perception Score represents how safe individuals feel  
- Safety Score combines both aspects for overall evaluation  

```python
df['Infrastructure_Score'] = (
    df['CCTV_Available'] +
    df['Security_in_Public_Place'] +
    df['Lighting_Condition']
)

df['Perception_Score'] = (
    df['Safe_Visiting_Alone'] +
    df['Avoid_Going_Out_At_Night'] +
    df['Experienced_Harassment']
)

df['Safety_Score'] = df['Infrastructure_Score'] + df['Perception_Score']
```

# Data Aggregation

Data is grouped by type of public place to compute average safety scores, enabling comparison across locations.

```python
place_data = df.groupby('Type_of_Public_Place').agg({
    'Infrastructure_Score':'mean',
    'Perception_Score':'mean',
    'Safety_Score':'mean'
}).reset_index()

print(place_data)
```

```python
place_data.to_csv("data/feature_data.csv", index=False)
```
