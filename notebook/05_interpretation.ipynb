# Interpretation and Ethical Analysis of Public Space Safety

This notebook focuses on interpreting the clustering results obtained from unsupervised learning models and analyzing their real-world implications. It also highlights ethical and societal considerations related to the use of such data.

## Interpretation of Clustering Results

The clustering techniques (K-Means, Hierarchical Clustering, and Ensemble Clustering) were used to group public places based on:

- Infrastructure Score (CCTV, security, lighting)
- Perception Score (women’s perceived safety)

### Key Observations:

- Public places with **high infrastructure and perception scores** fall into safer clusters.
- Some places have **good infrastructure but low perception**, indicating lack of trust or past negative experiences.
- Places with **low scores in both aspects** are identified as unsafe and require immediate attention.

### Cluster Meaning:

- Cluster 0 → Low Safety (poor infrastructure + negative perception)
- Cluster 1 → Moderate Safety (average conditions)
- Cluster 2 → High Safety (strong infrastructure + positive perception)

These clusters help categorize public spaces based on safety conditions.

```python
import pandas as pd
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler

place_data = pd.read_csv("data/clustered_data.csv")
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(
    place_data[['Infrastructure_Score','Perception_Score']]
)
```

## Model Evaluation

The Silhouette Score is used to evaluate clustering performance.

- A higher score indicates well-separated clusters
- A lower score suggests overlap between clusters

The obtained score shows that the clustering model forms reasonably distinct groups, making the analysis meaningful.

```python
score = silhouette_score(X, place_data['kmeans'])

print("Silhouette Score:", score)
```
