# Data Understanding and Preprocessing

This notebook focuses on loading the dataset, understanding its structure, and performing preprocessing steps to prepare the data for analysis.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Women_public_place_safety.csv")

print(df.head())
print(df.info())
```

## Data Cleaning

Inconsistent values such as different formats of the same category are standardized. Missing values are handled to ensure data quality.

```python
df['Lighting_Condition'] = df['Lighting_Condition'].replace('avg','Average')
df['Crowd_Level'] = df['Crowd_Level'].replace('1000-10k','1000-10000')

df['Mode_of_Transport'] = df['Mode_of_Transport'].fillna('Unknown')
```

## Encoding Categorical Variables

Categorical variables are converted into numerical form to make them suitable for machine learning algorithms.

```python
df['Safe_Visiting_Alone'] = df['Safe_Visiting_Alone'].map({'Yes':1,'No':0})
df['Avoid_Going_Out_At_Night'] = df['Avoid_Going_Out_At_Night'].map({'Yes':0,'No':1})
df['Experienced_Harassment'] = df['Experienced_Harassment'].map({'Yes':0,'No':1})
df['CCTV_Available'] = df['CCTV_Available'].map({'Yes':1,'No':0})
df['Security_in_Public_Place'] = df['Security_in_Public_Place'].map({'Yes':1,'No':0})

df['Lighting_Condition'] = df['Lighting_Condition'].map({
    'Poor':0,
    'Average':1,
    'Good':2
})
```
